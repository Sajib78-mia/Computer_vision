```python
import cv2
from matplotlib import pyplot as plt

# Load the image
image_path = '/content/Training_143373.jpg'
image = cv2.imread(image_path)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Task 3: Edge Detection for Facial Structure Analysis
def detect_edges(img, lower_threshold=100, upper_threshold=200):
    edges = cv2.Canny(img, lower_threshold, upper_threshold)
    return edges

# Detect edges
edges_image = detect_edges(gray_image)

# Display the original and edges-detected images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Edges Detected')
plt.imshow(edges_image, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
